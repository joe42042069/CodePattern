```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics

mu, sigma = 0, 0.5 # mean and standard deviation
samples = np.random.normal(mu, sigma, 10000) # generate 1000 samples

plt.hist(samples, bins=200, density=True, alpha=0.5, color='g')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Gaussian Distribution')

# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-2, 2, 0.01)
plt.plot(x_axis, norm.pdf(x_axis, mu, sigma))
plt.show()
```
